```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta_x=0.01
X=np.arange(-1,1+delta_x,delta_x)
len_x=len(X)
```

```python
delta_t=delta_x**2
T=np.arange(0,0.1+delta_t,delta_t)
len_t=len(T)
```

```python
Psi_sol=np.zeros([len_x,len_t])
```

```python
m=int((len_x-1)/2)
Psi_sol[m][0] = 1/(2*delta_x)
Psi_sol[m-1][0] = 1/(4*delta_x)
Psi_sol[m+1][0] = 1/(4*delta_x)
```

```python
plt.imshow(Psi_sol)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xticks([],[])
plt.yticks([],[])
plt.xlabel('t',size=20)
plt.ylabel('x',size=20)
plt.show()
```

```python
for j in range(len_t-1):
    for i in range(1,len_x-1):
        Psi_sol[i][j+1] = (Psi_sol[i+1][j]+Psi_sol[i-1][j])/2
```

```python
plt.plot(X,Psi_sol[::,0])
plt.plot(X,Psi_sol[::,1])
plt.plot(X,Psi_sol[::,2])
plt.show
```

```python
plt.imshow(Psi_sol,extent=(min(T), max(T), min(X), max(X)), aspect=0.04, vmin=0, vmax=3)
plt.xlabel(r'$t$', size=20)
plt.ylabel(r'$x$', size=20)
plt.colorbar()
plt.show()
```
