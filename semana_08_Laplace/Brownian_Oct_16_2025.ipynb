##Mov. Browniano en 2D

```python

delta_l = 0.01
posX = 0
posY = 0
X = [posX]
Y = [posY]

for j in range(10000):
    angulo = 2*np.pi*np.random.rand()
    posX += np.cos(angulo)*delta_l
    posY += np.sin(angulo)*delta_l
    X.append(posX)
    Y.append(posY)
```

```python
plt.plot(X,Y)
plt.axis('off')
plt.show
```
