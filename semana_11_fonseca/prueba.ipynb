```python
#Método de cuadrado medio (para extraer números pseudoaleatorios)
#14 -> 0196 -> 19
#19 -> 0361 -> 36
#36 -> 1296 -> 29
#29 -> 0841 -> 84
#84 -> 7056 -> 05
#05 -> 0025 -> 25
#02 -> 0004 -> 00
#00 -> 0000 -> 00
semilla = 7182
print(semilla)
def aleatorio():
    global semilla
    s1 = str(semilla**2)
    while (len(s1) != 8):
        s1 = '0' + s1
    semilla = int(s1[2:6])
    return semilla

for i in range(0,14):
    print(aleatorio())
```

```python
def rng(m=2**32, a=123456789, c=12345):
    rng.corriente = (a*rng.corriente +c) % m
    return rng.corriente/m

#¿Cómo se llama este algoritmo?
#Generador lineal congruencial
#¿Y para qué sirve?
#Para generar números pseudoaleatorios uniformemente distribuidos en el intervalo [0,1)
#¿Cómo funciona?
#Se basa en la fórmula recursiva Xn+1 = (a*Xn + c)
# mod m, donde X es el número pseudoaleatorio, y a, c, m son constantes elegidas adecuadamente.
#¿Cuáles son sus ventajas y desventajas?
#Ventajas: Es simple y rápido de implementar. Permite generar secuencias largas de números pseudoaleatorios.
#Desventajas: La calidad de los números generados depende de la elección de los parámetros a, c y m.
# Puede presentar patrones si no se eligen adecuadamente.
#¿Qué significa rng.corriente?
#Es una variable estática que almacena el valor actual del generador, permitiendo la generación del siguiente
# número pseudoaleatorio en la secuencia.
#¿Qué es una variable estática en este contexto?
#Es una variable que mantiene su valor entre llamadas a la función, permitiendo que el estado del generador
# se preserve.
```

```python
import numpy as np
import matplotlib.pyplot as plt

#------ Generación de números pseudoaleatorios ------#

def rng(m=2**32, a=1282047789, c=12345):
    rng.corriente = (a*rng.corriente +c) % m
    return rng.corriente/m
rng.corriente = 367282901

# Generar una secuencia de números pseudoaleatorios
n = 50000
muestras = np.array([rng() for _ in range(n)])

#------ Visualización de los números pseudoaleatorios ------#

plt.figure(figsize=(9, 3))
plt.plot(muestras[:500], lw=0.58)
plt.title('Serie - Primeras 500 muestras pseudoaleatorias generadas por rng()')
plt.xlabel('Índice de Muestra')
plt.ylabel('Valor Pseudoaleatorio')
plt.tight_layout()
#Embedding de imagen

#Graficar por pares seguidos
#Variable inicial
u = muestras[:-1] #Todos menos el último
v = muestras[1:] #Todos menos el primero
#Skater plot
#"Somos mutantes, pertenecemos a genosha" (Wolverine)
plt.figure(figsize=(6, 6))
plt.scatter(u, v, s=0.5)
plt.title('Diagrama de dispersión - Pares consecutivos de muestras pseudoaleatorias')
plt.xlabel('Muestra n')
plt.ylabel('Muestra n+1')
plt.tight_layout()
```

```python
#Ya se hizo un cuadrado lleno de números pseudoaleatorios
#¿Qué tal un cubo?
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d') #¿Por qué 111? - 1 fila, 1 columna, primer subplot
ax.scatter(muestras[:-2], muestras[1:-1], muestras[2:], s=0.5)
ax.set_title('Diagrama de dispersión 3D - Tripletas consecutivas de muestras pseudoaleatorias')
ax.set_xlabel('Muestra n')
ax.set_ylabel('Muestra n+1')
ax.set_zlabel('Muestra n+2')
plt.tight_layout() #¿Qué es esto? - Ajusta automáticamente los parámetros de la figura para que los elementos encajen bien.
```

```python
plt.figure(figsize=(5, 3))
bins = 50
plt.hist(muestras, bins=bins, density=True, alpha=0.7,label='empirico' ,color='blue', edgecolor='black')
x=np.linspace(0, 1, 200)
plt.plot(x, np.ones_like(x), 'r--', lw=2, label='teórico')
#plt.title('Histograma (%d bins) de muestras pseudoaleatorias generadas por rng()' % (bins,n))
#Arreglando title...
plt.title(f'Histograma ({bins} bins, {n} muestras)')
plt.xlabel('Valor Pseudoaleatorio')
plt.ylabel('Densidad de Probabilidad')
plt.tight_layout()
#Guardar imagen
plt.legend()
plt.savefig('histograma_rng.png', dpi=300)
plt.show()
```
